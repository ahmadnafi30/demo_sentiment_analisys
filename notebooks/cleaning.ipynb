```python
import re 

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)  
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'@[A-Za-z0-9_]+|#[A-Za-z0-9_]+|[^A-Za-z\s]', '', text)
    text = re.sub(r'\b[a-zA-Z](?:\s+[a-zA-Z])+\b', lambda m: m.group().replace(" ", ""), text)

    return text

def extract_hashtags(text):
    return re.findall(r'#\w+', text)

def clean_data(df):
    df.drop_duplicates(inplace=True)
    df.drop(columns=["conversation_id_str", "id_str", "location" ,"user_id_str", "username","tweet_url", "image_url", "lang", "in_reply_to_screen_name"], inplace=True)
    df['hashtags'] = df['full_text'].apply(extract_hashtags)
    df['cleaned_text'] = df['full_text'].apply(clean_text)
    df['cleaned_text'] = df['cleaned_text'].str.lower()
    return df
```

```python
df_cleaned = clean_data(df.copy())
```
